```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
npoints = 100
y=np.random.uniform(0,1,npoints)
print(y)
y_bar=np.average(y)
print("Average =", y_bar)
```

```python
lamda = 10
p=-lamda*np.log(1-y)
print(p)
p_a= np.average(p)
print("Average =",p_a)
p_s=np.std(p)
print("Standard Deviation =", p_s)
p_e= (p_s**2/npoints)**(1/2)
print("Error on the mean =",p_e)
```

```python
fig = plt.figure(figsize=(10,7))
plt.hist(p,density=True)
plt.xlabel("x")
plt.ylabel("y")
x = np.linspace(0,50,1000)
z = 1/lamda * np.e**(-x/lamda)
plt.plot(x,z)
```
